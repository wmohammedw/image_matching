```python
import cv2
import numpy as np
import jsonpickle
import requests
import matplotlib.pyplot as plt
```

```python
img1 = cv2.imread('house 1.jpg', cv2.IMREAD_GRAYSCALE)
```

```python
addr = "http://127.0.0.1:5000/isItSimilar/"
headers = {'content-type': 'image/jpeg'}

_, img_encoded = cv2.imencode('.jpg', img1)
requests.post(addr, data=img_encoded.tobytes(), headers=headers)
```

```python
info = requests.get("http://127.0.0.1:5000/isItSimilar2/")
deconded_info = jsonpickle.decode(info._content)
deconded_info
```

```python
plt.imshow(deconded_info['img'])
```
